```python
account_dfs_list = []

for account in accounts_list:
    
    account_dict = json.load(open(f'data/{account}.json'))
    
    account_df = pd.io.json.json_normalize(account_dict)
    
    account_dfs_list.append(account_df)
    
account_dfs_list
```

```python
type(account)
```

```python
all_accounts_df = pd.concat(account_dfs_list)
all_accounts_df
```

```python
accounts_df_final = all_accounts_df[['account_name', 'follower_count', 
                                     'following_count', 'post_count']]
```

```python
accounts_df_final.reset_index(inplace=True, drop=True)
```

```python
accounts_df_final.to_csv('account_info.csv')
```
